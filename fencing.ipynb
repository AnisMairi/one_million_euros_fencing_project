```python
def draw_optical_flow(prev, next):
    flow = cv2.calcOpticalFlowFarneback(prev, next, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    mag, ang = cv2.cartToPolar(flow[...,0], flow[...,1])
    hsv = np.zeros_like(cv2.cvtColor(prev, cv2.COLOR_GRAY2BGR))
    hsv[..., 1] = 255
    hsv[..., 0] = ang * 180 / np.pi / 2
    hsv[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
```

```python
import cv2
import numpy as np
import math
from ultralytics import YOLO
from collections import deque
import mediapipe as mp
import torch
import torchvision.transforms as T
from torchvision.transforms.functional import to_pil_image
from midas.dpt_depth import DPTDepthModel

# Load YOLO model
yolo = YOLO('yolov8n.pt')  # use yolov8s.pt for better accuracy

# Load MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=False, model_complexity=1, min_detection_confidence=0.5)

# Load MiDaS depth model
midas = torch.hub.load("intel-isl/MiDaS", "DPT_Large")
midas.eval()
midas_transforms = torch.hub.load("intel-isl/MiDaS", "transforms").dpt_transform

# Setup video I/O
cap = cv2.VideoCapture('fencing_match.mp4')
if not cap.isOpened():
    raise FileNotFoundError("Video not found or unreadable.")

w, h = int(cap.get(3)), int(cap.get(4))
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('fencing_4views_scene.mp4', fourcc, fps, (w * 2, h * 2))

# Pose indices
WRIST = 16  # RIGHT_WRIST
ELBOW = 14  # RIGHT_ELBOW
SHOULDER = 12  # RIGHT_SHOULDER

prev_gray = None

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    time_sec = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000.0

    # ---- YOLO View ----
    yolo_view = frame.copy()
    yolo_results = yolo(yolo_view, classes=[0], verbose=False)
    boxes = [list(map(int, box.xyxy[0])) for r in yolo_results for box in r.boxes if float(box.conf[0]) > 0.3]
    for x1, y1, x2, y2 in boxes:
        cv2.rectangle(yolo_view, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # ---- Pose View ----
    pose_view = frame.copy()
    pose_result = pose.process(frame_rgb)
    if pose_result.pose_landmarks:
        mp.solutions.drawing_utils.draw_landmarks(
            pose_view, pose_result.pose_landmarks, mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp.solutions.drawing_styles.get_default_pose_landmarks_style())

    # ---- Depth View ----
    input_tensor = midas_transforms(frame_rgb).to('cpu')
    with torch.no_grad():
        prediction = midas(input_tensor)[0]
    depth = cv2.normalize(prediction.numpy(), None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
    depth_view = cv2.applyColorMap(depth, cv2.COLORMAP_MAGMA)
    depth_view = cv2.resize(depth_view, (w, h))

    # ---- Scene Flow View (Optical Flow) ----
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    if prev_gray is None:
        scene_flow_view = np.zeros_like(frame)
    else:
        flow = cv2.calcOpticalFlowFarneback(prev_gray, gray,
                                            None, 0.5, 3, 15, 3, 5, 1.2, 0)
        mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])
        hsv = np.zeros_like(frame)
        hsv[..., 1] = 255
        hsv[..., 0] = ang * 180 / np.pi / 2
        hsv[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
        scene_flow_view = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    prev_gray = gray

    # Add labels to each view
    cv2.putText(yolo_view, 'YOLO Bounding Boxes', (10, 30), 0, 1, (255, 255, 255), 2)
    cv2.putText(pose_view, 'Pose Estimation', (10, 30), 0, 1, (255, 255, 255), 2)
    cv2.putText(depth_view, 'Depth Estimation', (10, 30), 0, 1, (255, 255, 255), 2)
    cv2.putText(scene_flow_view, 'Scene Flow (Optical Flow)', (10, 30), 0, 1, (255, 255, 255), 2)

    # Combine views into 2x2 layout
    top = np.hstack((yolo_view, pose_view))
    bottom = np.hstack((depth_view, scene_flow_view))
    combined = np.vstack((top, bottom))

    out.write(combined)

cap.release()
out.release()
cv2.destroyAllWindows()
print("✅ Output video saved as fencing_4views_scene.mp4")
```

```python
import cv2
import numpy as np
import math
from ultralytics import YOLO
from collections import deque
import mediapipe as mp
import torch
import torchvision.transforms as T
from torchvision.transforms.functional import to_pil_image
from midas.dpt_depth import DPTDepthModel
from openpifpaf import Predictor  # For 3D pose

# Load YOLO model
yolo = YOLO('yolov8n.pt')  # use yolov8s.pt for better accuracy

# Load MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=False, model_complexity=1, min_detection_confidence=0.5)

# Load MiDaS depth model
midas = torch.hub.load("intel-isl/MiDaS", "DPT_Large")
midas.eval()
midas_transforms = torch.hub.load("intel-isl/MiDaS", "transforms").dpt_transform

# Load 3D pose predictor (OpenPifPaf or similar)
predictor = Predictor(checkpoint='shufflenetv2k16')

# Setup video I/O
cap = cv2.VideoCapture('fencing_match.mp4')
if not cap.isOpened():
    raise FileNotFoundError("Video not found or unreadable.")

w, h = int(cap.get(3)), int(cap.get(4))
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('fencing_5views_scene.mp4', fourcc, fps, (w * 3, h * 2))

# Pose indices
WRIST = 16  # RIGHT_WRIST
ELBOW = 14  # RIGHT_ELBOW
SHOULDER = 12  # RIGHT_SHOULDER

prev_gray = None

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    time_sec = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000.0

    # ---- YOLO View ----
    yolo_view = frame.copy()
    yolo_results = yolo(yolo_view, classes=[0], verbose=False)
    boxes = [list(map(int, box.xyxy[0])) for r in yolo_results for box in r.boxes if float(box.conf[0]) > 0.3]
    for x1, y1, x2, y2 in boxes:
        cv2.rectangle(yolo_view, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # ---- Pose View ----
    pose_view = frame.copy()
    pose_result = pose.process(frame_rgb)
    if pose_result.pose_landmarks:
        mp.solutions.drawing_utils.draw_landmarks(
            pose_view, pose_result.pose_landmarks, mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp.solutions.drawing_styles.get_default_pose_landmarks_style())

    # ---- Depth View ----
    input_tensor = midas_transforms(frame_rgb).to('cpu')
    with torch.no_grad():
        prediction = midas(input_tensor)[0]
    depth = cv2.normalize(prediction.numpy(), None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
    depth_view = cv2.applyColorMap(depth, cv2.COLORMAP_MAGMA)
    depth_view = cv2.resize(depth_view, (w, h))

    # ---- Scene Flow View (Optical Flow) ----
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    if prev_gray is None:
        scene_flow_view = np.zeros_like(frame)
    else:
        flow = cv2.calcOpticalFlowFarneback(prev_gray, gray,
                                            None, 0.5, 3, 15, 3, 5, 1.2, 0)
        mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])
        hsv = np.zeros_like(frame)
        hsv[..., 1] = 255
        hsv[..., 0] = ang * 180 / np.pi / 2
        hsv[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
        scene_flow_view = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    prev_gray = gray

    # ---- 3D Pose View ----
    pose3d_view = frame.copy()
    predictions = predictor.numpy_image(frame_rgb)
    for pred in predictions:
        for kpt in pred.data:
            x, y, conf = int(kpt[0]), int(kpt[1]), kpt[2]
            if conf > 0.3:
                cv2.circle(pose3d_view, (x, y), 3, (0, 255, 255), -1)

    # Add labels to each view
    cv2.putText(yolo_view, 'YOLO Bounding Boxes', (10, 30), 0, 1, (255, 255, 255), 2)
    cv2.putText(pose_view, 'Pose Estimation', (10, 30), 0, 1, (255, 255, 255), 2)
    cv2.putText(depth_view, 'Depth Estimation', (10, 30), 0, 1, (255, 255, 255), 2)
    cv2.putText(scene_flow_view, 'Scene Flow (Optical Flow)', (10, 30), 0, 1, (255, 255, 255), 2)
    cv2.putText(pose3d_view, '3D Pose Estimation', (10, 30), 0, 1, (255, 255, 255), 2)

    # Resize all views to same dimensions
    yolo_view = cv2.resize(yolo_view, (w, h))
    pose_view = cv2.resize(pose_view, (w, h))
    scene_flow_view = cv2.resize(scene_flow_view, (w, h))
    pose3d_view = cv2.resize(pose3d_view, (w, h))

    # Combine views into 3x2 layout (5 views + 1 black filler)
    black = np.zeros_like(frame)
    top_row = np.hstack((yolo_view, pose_view, depth_view))
    bottom_row = np.hstack((scene_flow_view, pose3d_view, black))
    combined = np.vstack((top_row, bottom_row))

    out.write(combined)

cap.release()
out.release()
cv2.destroyAllWindows()
print("✅ Output video saved as fencing_5views_scene.mp4")
```
